```python
import pandas as pd
import numpy as np
from sklearn import svm
```

```python
tcs_day = pd.read_csv('D:\Data-Analysis\Stock Market/TCS_Daily.csv')
tcs_day
```

```python
run_avg = np.array((tcs_day['Open Price'] + tcs_day['Close Price'] + tcs_day['High Price'] + tcs_day['Low Price'])/4)
trns = np.array(tcs_day['Total Turnover (Rs.)'])

print(run_avg)
```

```python
m, k = 1400, 1700

x=np.transpose([run_avg, trns])
y = []

for i in range(len(run_avg)):
    if run_avg[i]<m:
        y.append(0)
    elif run_avg[i]>=m and run_avg[i]<=k:
        y.append(1)
    else:
        y.append(2)
        
print(y)
```

```python
x_train = x[:-200]
y_train = y[:-200]
x_test = x[-200:]
y_test = y[-200:]
```

```python
svc = svm.SVC(kernel='poly')
svc.fit(x_train, y_train)
```

```python
svc.score(x_test, y_test)
```
