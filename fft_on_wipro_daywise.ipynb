```
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import pandas as pd
import numpy as np
```

https://docs.scipy.org/doc/scipy/reference/tutorial/fftpack.html#dct-and-idct

```
data = pd.read_csv('Wipro.csv')
```

```
data.shape
```

```
plt.plot(data['Close Price'][0:200])
```

```
yf = fft(data['Close Price'][0:200])
plt.plot(yf)
plt.grid()
plt.show()
```

```
yf = fft(data['Close Price']/data['Close Price'].max())
plt.plot(yf)
plt.grid()
plt.show()
```
