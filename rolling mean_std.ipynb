```python
import pandas as pd
import matplotlib.pyplot as plt
data_wipro_day = pd.read_csv("C:/Users/Anirban/Downloads/dataset_daywise/wipro_daywise.csv",usecols = ['Date','Open Price'])
data_wipro_day = data_wipro_day.head(50)
```

```python
# This decision of buying & selling point was made on 'Adj. close price' on the reference.
# But due to absence of that feature, I used 'Open price'----
```

```python
rm = data_wipro_day['Open Price'].rolling(10).mean()
rstd = data_wipro_day['Open Price'].rolling(10).std()
upper_band = rm+2*rstd
lower_band = rm-2*rstd
```

```python
import pylab
plt.figure(figsize=(10,10))
ax = plt.subplot(111)
plt.plot(data_wipro_day['Open Price'], label="Open price")
plt.plot(rm,label ="rolling mean")
plt.plot(upper_band,label = "upper band")
plt.plot(lower_band,label="lower band")
pylab.legend(loc='upper left')
plt.grid()
plt.annotate('Buying point', xy=(21.5, 273), 
            arrowprops=dict(facecolor='black', shrink=0.5),
            )
plt.annotate('Selling point', xy=(26, 288), 
            arrowprops=dict(facecolor='black', shrink=0.5),
            )
plt.show()
```
